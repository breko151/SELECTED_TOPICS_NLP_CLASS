```python
# Importar las bibliotecas necesarias.
import spacy
# Inicialización de spacy.
nlp = spacy.load("en_core_web_sm")
# Función de normalización.
def normalize(text):
    doc = nlp(text)
    words = [[t.orth_, t.pos_] for t in doc if not t.is_punct | t.is_stop]
    lexical_tokens = [(t[0].lower(), t[1]) for t in words if len(t[0]) > 3 and t[0].isalpha()]    
    return lexical_tokens
```

```python
import phunspell
corrector_hunspell = phunspell.Phunspell('en_US')
```

```python
def corrector(tokens):
    new_dict = dict()
    contador = 0
    for token, value in tokens.items():
        contador += 1
        print("Token:", contador)
        if token[1] == 'PROPN' or token[1] == 'NOUN':
            print("\tPasa la condicion")
            sugerencias = list(corrector_hunspell.suggest(token[0]))
            if sugerencias:
                new_dict[(sugerencias[0], token[1])] = value
            else:
                new_dict[token] = value
        else:
            new_dict[token] = value
        
    return new_dict
```

```python
path = "./nlp-text-exercise.txt"
with open(path, 'r', encoding="utf-8") as archivo:
    # Lee todo el contenido del archivo en una variable
    contenido = archivo.read()
```

```python
arreglo = normalize(contenido)
```

```python
vocabulary = set(arreglo)
```

```python
voc_dict = dict()
for token in vocabulary:
    voc_dict[token] = arreglo.count(token)
```

```python
arreglo_correcto = corrector(voc_dict)
```

```python
def lemmatizacion(dict_values):
    new_dict = dict()
    for k, v in dict_values.items():
        doc = nlp(k[0])
        token = [(t.lemma_, t.pos_) for t in doc][0]
        new_dict[token] = v
    #texto = " ".join(arreglo)
    #doc = nlp(texto)
    #lemmas_tokens = [(t.lemma_, t.pos_) for t in doc]  
    return new_dict
```

```python
arreglo_lemmas = lemmatizacion(arreglo_correcto)
```

```python
import contractions
```

```python
def expand_text(dict_values):
    new_dict = dict()
    for token, v in dict_values.items():
        new_token = (contractions.fix(token[0]), token[1])
        new_dict[new_token] = v
    return new_dict
```

```python
texto_expand = expand_text(arreglo_lemmas)
```

```python
texto_expand
```

```python
pos = list(set([k[1] for k in texto_expand.keys()]))
dic_pos = {p: 0 for p in pos}
```

```python
for k, v in texto_expand.items():
    for p in pos:
        if p == k[1]:
            dic_pos[p] = dic_pos[p] + v
```
