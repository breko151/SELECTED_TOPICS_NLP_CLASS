```python
# Importar las bibliotecas necesarias.
import spacy
# Inicialización de spacy.
nlp = spacy.load("en_core_web_sm")
```

```python
path_1 = './human_text.txt'
path_2 = './robot_text.txt'
```

```python
with open(path_1, 'r', encoding="utf-8") as archivo:
    # Lee todo el contenido del archivo en una variable
    contenido_1 = archivo.read()
```

```python
with open(path_2, 'r', encoding="utf-8") as archivo:
    # Lee todo el contenido del archivo en una variable
    contenido_2 = archivo.read()
```

```python
def normalize(text):
    doc = nlp(text)
    words = [[t.lemma_, t.pos_] for t in doc if not t.is_punct | t.is_stop]
    lexical_tokens = [(t[0].lower(), t[1]) for t in words if len(t[0]) > 3 and t[0].isalpha()]    
    return lexical_tokens
```

```python
contenido_total =  contenido_1 + "\n" + contenido_2
```

```python
longitud_texto =  len(contenido_total)
nlp.max_length =  longitud_texto + 100000
```

```python
# './human_text.txt'
arreglo_1 = normalize(contenido_1)
```

```python
# './human_text.txt'
arreglo_2 = normalize(contenido_2)
```

```python
arreglo_juntos = arreglo_1 + arreglo_2
```

```python
vocabulary = set(arreglo_juntos)
```

```python
# './human_text.txt'
voc_dict_1 = dict()
for token in vocabulary:
    voc_dict_1[token] = arreglo_1.count(token)
```

```python
# './robot_text.txt'
voc_dict_2 = dict()
for token in vocabulary:
    voc_dict_2[token] = arreglo_2.count(token)
```
