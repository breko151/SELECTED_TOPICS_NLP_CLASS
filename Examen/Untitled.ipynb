```python
# Importar las bibliotecas necesarias.
import spacy
# InicializaciÃ³n de spacy.
nlp = spacy.load("en_core_web_sm")
import phunspell
corrector_hunspell = phunspell.Phunspell('en_US')
```

```python
path_1 = './human_text.txt'
path_2 = './robot_text.txt'
```

```python
with open(path_1, 'r', encoding="utf-8") as archivo:
    # Lee todo el contenido del archivo en una variable
    contenido_1 = archivo.read()
```

```python
with open(path_2, 'r', encoding="utf-8") as archivo:
    # Lee todo el contenido del archivo en una variable
    contenido_2 = archivo.read()
```

```python
contenido_total =  contenido_1 + "\n" + contenido_2
```

```python
longitud_texto =  len(contenido_total)
nlp.max_length =  longitud_texto + 100000
```

```python
def normalize(text):
    doc = nlp(text)
    words = [[t.orth_, t.pos_] for t in doc if not t.is_punct | t.is_stop]
    lexical_tokens = [(t[0].lower(), t[1]) for t in words if len(t[0]) > 3 and t[0].isalpha()]    
    return lexical_tokens
```

```python
def corrector(tokens):
    new_dict = dict()
    contador = 0
    for token, value in tokens.items():
        contador += 1
        print("Token:", contador)
        if token[1] == 'PROPN' or token[1] == 'NOUN':
            print("\tPasa la condicion")
            sugerencias = list(corrector_hunspell.suggest(token[0]))
            if sugerencias:
                new_dict[(sugerencias[0], token[1])] = value
            else:
                new_dict[token] = value
        else:
            new_dict[token] = value
        
    return new_dict
```

```python
def lemmatizacion(dict_values):
    new_dict = dict()
    for k, v in dict_values.items():
        doc = nlp(k[0])
        token = [(t.lemma_, t.pos_) for t in doc][0]
        new_dict[token] = v
    #texto = " ".join(arreglo)
    #doc = nlp(texto)
    #lemmas_tokens = [(t.lemma_, t.pos_) for t in doc]  
    return new_dict
```

```python
arreglo = normalize(contenido_total)
```

```python
vocabulary = set(arreglo)
```

```python
voc_dict = dict()
for token in vocabulary:
    voc_dict[token] = arreglo.count(token)
```

```python
arreglo_correcto = corrector(voc_dict)
```

```python
# Bolsa de palabras total
arreglo_lemmas = lemmatizacion(arreglo_correcto)
```
