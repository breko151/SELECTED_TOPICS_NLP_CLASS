```python
def fix(texto):
    expanded_words = []
    for word in texto.lower().split():
        if word in CONTRACTION_MAP:
            new_word = CONTRACTION_MAP[word]
        else:
            new_word = word
        expanded_words.append(new_word)
    expanded_text = ' '.join(expanded_words)
    print("Original Text: " + texto)
    print("Expanded Text: " + expanded_text)
    return expanded_text
```

```python
texto =  "I will fix you"
tokens = texto.lower().split()
bigrams = [[tokens[i], tokens[i + 1]] for i in range(len(tokens) - 1)]
bigrams_text = [' '.join(bigram) for bigram in bigrams]
EXPAND_MAP = {valor: clave for clave, valor in CONTRACTION_MAP.items()}
expanded_words = list()
for word in bigrams_text:
        if word in EXPAND_MAP:
            new_word = EXPAND_MAP[word]
        else:
            new_word = word
        expanded_words.append(new_word)
cleaning_text = list()
for part in expanded_words:
    list_part = part.split()
    if len(list_part) == 1:
        cleaning_text.append(list_part[0])
    else:
        cleaning_text.append(list_part[1])
contracted = " ".join(cleaning_text)
print(contracted)
```
